```python
#Import multiple data file
import glob
import os
path = 'C:/Users/Pham Anh Tuan/Documents/DataSciencePython/gas_energy/Gas/'  # use your path
all_files_gas = glob.glob(path + '*.csv')
all_files_gas
```

```python
#li_energy = [pd.read_csv(filename, index_col = None, header = 0) for filename in all_files_energy]
li_gas = []
for csv in all_files_energy:
    frame = pd.read_csv(csv)
    frame['filename'] = os.path.basename(csv)
    li_gas.append(frame)
li_gas[0][0:5]
```

```python
df_gas = pd.concat(li_gas, axis = 0, ignore_index = True)
df_gas.head()
```
